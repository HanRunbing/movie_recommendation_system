```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')
users = pd.read_csv('clearn_users.csv')
```

```python
user_item_matrix = ratings.pivot(index = 'user_id', columns = 'movie_id', values = 'rating')
```

```python
user_item_matrix.head()
```

# user-based collaborative filtering

```python
user_item_matrix.loc[1, :].mean()
```

```python
# normalize the user-item matrix along users
user_mean = user_item_matrix.mean(axis=1)
normalized_matrix = user_item_matrix.sub(user_mean, axis=0)
```

```python
normalized_matrix.head()
```

```python
# calculate cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
user_sim = cosine_similarity(normalized_matrix.fillna(0).T)
```

```python
user_sim
```

# Item-based collaborative filtering

```python
item_mean = user_item_matrix.mean(axis=0)
normalized_item_matrix = user_item_matrix.sub(item_mean, axis=1)
```

```python
normalized_item_matrix.head()
```

```python
item_sim = cosine_similarity(normalized_item_matrix.fillna(0).T)
```
